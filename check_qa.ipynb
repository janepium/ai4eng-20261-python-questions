# Q/A solution check

Verificación de la función `preparar_datos` usando un generador de casos de uso.

```python
import importlib.util
import sys
import numpy as np
import pandas as pd
```

```python
# Cargar generador
spec_q = importlib.util.spec_from_file_location(
    'q',
    'myquestions/question-0001-usecase-generator.py'
)
q = importlib.util.module_from_spec(spec_q)
sys.modules['q'] = q
spec_q.loader.exec_module(q)

print('Generador cargado correctamente.')
```

```python
# Cargar respuesta
spec_a = importlib.util.spec_from_file_location(
    'a',
    'myanswers/answer-0001.py'
)
a = importlib.util.module_from_spec(spec_a)
sys.modules['a'] = a
spec_a.loader.exec_module(a)

print('Respuesta cargada correctamente.')
```

## Generar caso de prueba

```python
args, expected_output = q.generar_caso_de_uso_preparar_datos()

print('=== INPUT ===')
print('target_col:', args['target_col'])
display(args['df'])
```

## Ejecutar solución

```python
my_output = a.preparar_datos(**args)
print('Función ejecutada.')
```

## Comparación

```python
X_expected, y_expected = expected_output
X_my, y_my = my_output

print('¿X coincide?', np.allclose(X_expected, X_my))
print('¿y coincide?', np.array_equal(y_expected, y_my))
```

```python
# Pruebas múltiples
resultados = []

for i in range(5):
    args, expected_output = q.generar_caso_de_uso_preparar_datos()
    my_output = a.preparar_datos(**args)

    X_expected, y_expected = expected_output
    X_my, y_my = my_output

    resultados.append(np.allclose(X_expected, X_my) and np.array_equal(y_expected, y_my))

print('Todas correctas:', all(resultados))
```
